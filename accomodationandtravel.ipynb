# I - Data reading

In this section, we'll do several datasets retrieving

In the cell bellow, we perform a scanner in the list of datasets and read the ones that we need.

```python
# Define the datasets list
data_files = [
    "requisicaopassagem.csv",
    "requisicaohospedagem.csv",
]

# Define the dataSets path
data_path = "dataSets/"

# Declare the variable to keep all the data combined
data = {}

# Iterate through all dataSets
for f in data_files:
    # Perform the individual import of each dataSet separated by ";"
    d = pd.read_csv((data_path + "{0}").format(f), delimiter=";")
    
    # Define the key_name according to the dataSet filename
    key_name = f.replace(".csv", "")
    
    # Apply the data to the key
    data[key_name] = d
    
    # Print the imported dataSet shape (rows x cols)
    print(d.shape)
    
# Note: To access the dataSet within data{} variable
# you need to perform the following: data['key_name'].head()
# where key_name = the real name of the dataSet, i.e: requisicaohospedagem
```

## 2.1) Accomodation requests


This dataset includes all the accomodation requests from UFRN since 2006.

- numero - request ID;
- ano - year the request was made;
- proponente - name of the person who made the request; 
- proposto - name of the person who asked for the request;
- tipo_proposto - category of person who asked for the request(employee of the university or external person);
- data_cadastro - date the request was registered;
- data_entrada - date of the check-in in the accomodation facility; (not mandatory)
- data_saida - date of the check-out in the accomodation facility; (not mandatory)
- convenio - ?; (not mandatory)
- id_unidade_requisitante - ID of the department that requested the accomodation; 
- nome_unidade_requisitante - name of the department that requested the accomodation; 
- id_unidade_custo - ID of the department that will afford the request;
- nome_unidade_custo - name of the department that will afford the request;
- valor - amount of money of the request;
- status - status of the request; (not mandatory)

```python
#Preview of the requisicaohospedagem dataset
accomodation_data = data["requisicaohospedagem"]
print(accomodation_data)
```

```python
#We can see with the method info, that few data is missing, which is good for our analysis
data["requisicaohospedagem"].info()
```

```python
# Creating a new dataframe only with the year and value of the requests
reqhosp_anovalor = data["requisicaohospedagem"].loc[:, ["ano", "valor"]]
```

```python
#Sum of the amout of money spent in every year with accomodation
totalhosp = reqhosp_anovalor.groupby('ano')['valor'].sum()
totalhosp
```

```python
#Plotting the amount of money spend at UFRN with accomodation for the last 12 years
years = [2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]
hosptotal = [273414, 227893, 466782, 634016, 741168, 724297, 965706, 1219232, 1187114, 1010665, 1026578, 831617, 610672]

hosptotal_scaled = [value / 1000 for value in hosptotal]
plt.rcParams['figure.figsize'] = (11,7)
plt.plot(years, hosptotal_scaled, 'b')
plt.scatter(years, hosptotal_scaled, color='red')
plt.title('Gastos com hospedagem na UFRN de 2006 a 2018')
plt.xlabel("Ano")
plt.ylabel("Gastos (em milhares de reais)")
plt.xticks(years)
plt.show()
```

```python
#Plotting a histogram with the amount of accomodation requests

plt.hist(data["requisicaohospedagem"]['ano'], orientation='vertical')
plt.title('Historiograma da quantidade de requisições de hospedagem em função dos anos')
```

```python
#Names of all the departments (and programs) from UFRN that has requested for accomodation budget

totaldeptos = data["requisicaohospedagem"]['nome_unidade_requisitante'].unique()
print(totaldeptos)
```

```python
#Number (and sum of the amount of money) of each status of accomodation request
data["requisicaohospedagem"].groupby('status').agg({'valor': ['count','sum']})
```

```python
#Sum of the amout of money spent in every year with accomodation
#We can see that some status have a lot of NaN values. It happens because this column of the dataset is not mandatory
#In the following cell, we'll treat these data
data["requisicaohospedagem"].groupby(['ano','status'],as_index = False).sum().pivot('ano','status')
```

```python
#Treating the missing values
data["requisicaohospedagem"].groupby(['ano','status'],as_index = False).sum().pivot('ano','status').fillna(0)
```

```python
#All columns from the requisicaopassagem dataset

travel_data = data["requisicaopassagem"]
travel_data.info()
```

```python
#We can see from the cell above that the column 'observacaoes' has a lot of missing data, less the half oh the total amount.
#Also the colum 'data_autorizacao_chefe' has over 10 thousand missing data pieces.
#We'll work on a script below to drop these columns.
#The columns with fewer missing values will also be treated.

def remove_cols(input_df,columns_todrop):
    for dropped_column in columns_todrop:
        if dropped_column in input_df:
            input_df = input_df.drop([dropped_column], axis=1)
    return input_df


travel_dropped = ['observacoes', 'data_autorizacao_chefe']
travel_data = remove_cols(travel_data, travel_dropped)
travel_data.head()
```
